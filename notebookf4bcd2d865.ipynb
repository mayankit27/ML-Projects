Iris Flower Classification Project

```python
#Step-1:- Import Necessary libraries 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
import numpy as np


#Step-2:- load the dataset of iris flower
data=load_iris()
#print(data)
x=data.data
print(x[:30])
y=data.target
print(y[:30])

print('\nFeature Names:- ',data.feature_names)
print("\nTarget Names:- ",data.target_names)

#Step-3:-  Split the data into training and testing data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#Step-4:- Create the model and train the model by providing trained data
model=LogisticRegression(max_iter=200)
print(model)
model.fit(x_train,y_train)

#Step-5:- Predict the result based on tested data 
calculated_test_data=model.predict(x_test)

#Step-6:- Calculate the accuracy to know how good model is trained by comparing calculated data with actual test data
print('\nCalculated data :- ',calculated_test_data)
print("\nActual data:- ",y_test)
accuracy=accuracy_score(calculated_test_data,y_test)
print("\nAccuracy Score:- ",accuracy)
```
